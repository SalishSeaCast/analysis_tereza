```python
Jdf2013 = pd.read_excel('./Chldat/2013_Chla_JuanDeFuca_weekly_stats.xlsx')
Jdf2014 = pd.read_excel('./Chldat/2014_Chla_JuanDeFuca_weekly_stats.xlsx')
Jdf2015 = pd.read_excel('./Chldat/2015_Chla_JuanDeFuca_weekly_stats.xlsx')
Jdf2016 = pd.read_excel('./Chldat/2016_Chla_JuanDeFuca_weekly_stats.xlsx')

NS2013 = pd.read_excel('./Chldat/2013_Chla_North_weekly_stats.xlsx')
NS2014 = pd.read_excel('./Chldat/2014_Chla_North_weekly_stats.xlsx')
NS2015 = pd.read_excel('./Chldat/2015_Chla_North_weekly_stats.xlsx')
NS2016 = pd.read_excel('./Chldat/2016_Chla_North_weekly_stats.xlsx')

CS2013 = pd.read_excel('./Chldat/2013_Chla_Central_weekly_stats.xlsx')
CS2014 = pd.read_excel('./Chldat/2014_Chla_Central_weekly_stats.xlsx')
CS2015 = pd.read_excel('./Chldat/2015_Chla_Central_weekly_stats.xlsx')
CS2016 = pd.read_excel('./Chldat/2016_Chla_Central_weekly_stats.xlsx')
```

```python
fact = 0.55
fig, axs = plt.subplots(4,1, figsize=(20*fact, 14*fact), facecolor='w', edgecolor='k')
axs = axs.ravel()


dfs = [Jdf2013,Jdf2014,Jdf2015,Jdf2016]
for i in range(0,4):
    tdf = dfs[i]
    axs[i].plot((tdf['Week_Start_Date']+dt.timedelta(days=3)),\
             tdf['Interpolated_Median_Chla'], linestyle = '-', linewidth = 2, color = 'darkblue', \
                label = 'JdF interpolated median')
    axs[i].plot((tdf['Week_Start_Date'][0:44]+dt.timedelta(days=3)),\
             tdf['Interpolated_Mean_Chla'][0:44], linestyle = ':', linewidth = 2, color = 'darkblue', \
                label = 'interpolated mean')
    
dfs = [CS2013,CS2014,CS2015,CS2016]
for i in range(0,4):
    tdf = dfs[i]
    axs[i].plot((tdf['Week_Start_Date']+dt.timedelta(days=3)),\
             tdf['Interpolated_Median_Chla'], linestyle = '-', linewidth = 2, color = 'skyblue', \
                label = 'CS interpolated median')
    axs[i].plot((tdf['Week_Start_Date'][0:44]+dt.timedelta(days=3)),\
             tdf['Interpolated_Mean_Chla'][0:44], linestyle = ':', linewidth = 2, color = 'skyblue', \
                label = 'interpolated mean')
    
dfs = [NS2013,NS2014,NS2015,NS2016]
for i in range(0,4):
    tdf = dfs[i]
    axs[i].plot((tdf['Week_Start_Date']+dt.timedelta(days=3)),\
             tdf['Interpolated_Median_Chla'], linestyle = '-', linewidth = 2, color = 'red', \
                label = 'NS interpolated median')
    axs[i].plot((tdf['Week_Start_Date'][0:44]+dt.timedelta(days=3)),\
             tdf['Interpolated_Mean_Chla'][0:44], linestyle = ':', linewidth = 2, color = 'red', \
                label = 'interpolated mean')
    
axs[0].set_xlim(['2013-03-01','2013-11-30'])
axs[1].set_xlim(['2014-03-01','2014-11-30'])
axs[2].set_xlim(['2015-03-01','2015-11-30'])
axs[3].set_xlim(['2016-03-01','2016-11-30'])


plt.legend(bbox_to_anchor = [1.5, 1.2])
axs[0].set_title('Satellite chlorophyll means/meadians by region (Suchy/Costa data)')
```
