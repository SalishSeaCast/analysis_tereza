### calculate monthly by-depth averages - 7 x 40 x 12
### save all as pkl - also divide by size of cell to get concentration


```python
##ANOM DIC 
anomDIC_2050_flux_stor = np.zeros([7,40,12])

hinds = [trans_anomDIC_2050_2013_mol_per_gridcell,trans_anomDIC_2050_2014_mol_per_gridcell,trans_anomDIC_2050_2015_mol_per_gridcell,
         trans_anomDIC_2050_2016_mol_per_gridcell,trans_anomDIC_2050_2017_mol_per_gridcell,
        trans_anomDIC_2050_2018_mol_per_gridcell,trans_anomDIC_2050_2019_mol_per_gridcell]

for yr in range(0,7):
    hind = hinds[yr]; hind[hind == 0] = np.nan
    for m in range(0,12):
        hind_month = np.nanmean(hind[m*30:m*30+30,:,:],axis =0) 
        anomDIC_2050_flux_stor[yr,:,m] = np.nansum(hind_month, axis = 1)
            
pickle.dump(anomDIC_2050_flux_stor, open("anomDIC_2050_flux_stor.pkl", 'wb'))
#DIC
DIC_flux_stor = np.zeros([7,40,12])

hinds = [trans_DIC_2013_mol_per_gridcell,trans_DIC_2014_mol_per_gridcell,trans_DIC_2015_mol_per_gridcell,
         trans_DIC_2016_mol_per_gridcell,trans_DIC_2017_mol_per_gridcell,
        trans_DIC_2018_mol_per_gridcell,trans_DIC_2019_mol_per_gridcell]

for yr in range(0,7):
    hind = hinds[yr]; hind[hind == 0] = np.nan
    for m in range(0,12):
        hind_month = np.nanmean(hind[m*30:m*30+30,:,:],axis =0) 
        DIC_flux_stor[yr,:,m] = np.nansum(hind_month, axis = 1)
            
pickle.dump(DIC_flux_stor, open("DIC_flux_stor.pkl", 'wb'))
#phys
phys_flux_stor = np.zeros([7,40,12])

hinds = [trans_phys_2013_kg_per_gridcell,trans_phys_2014_kg_per_gridcell,trans_phys_2015_kg_per_gridcell,
         trans_phys_2016_kg_per_gridcell,trans_phys_2017_kg_per_gridcell,
        trans_phys_2018_kg_per_gridcell,trans_phys_2019_kg_per_gridcell]

for yr in range(0,7):
    hind = hinds[yr]; hind[hind == 0] = np.nan
    for m in range(0,12):
        hind_month = np.nanmean(hind[m*30:m*30+30,:,:],axis =0) 
        phys_flux_stor[yr,:,m] = np.nansum(hind_month, axis = 1)
            
pickle.dump(phys_flux_stor, open("phys_flux_stor.pkl", 'wb'))

anomsal_30pt0_flux_stor = np.zeros([7,40,12])

hinds = [trans_anomsal_30pt0_2013_kg_per_gridcell,trans_anomsal_30pt0_2014_kg_per_gridcell,trans_anomsal_30pt0_2015_kg_per_gridcell,
         trans_anomsal_30pt0_2016_kg_per_gridcell,trans_anomsal_30pt0_2017_kg_per_gridcell,
        trans_anomsal_30pt0_2018_kg_per_gridcell,trans_anomsal_30pt0_2019_kg_per_gridcell]

for yr in range(0,7):
    hind = hinds[yr]; hind[hind == 0] = np.nan
    for m in range(0,12):
        hind_month = np.nanmean(hind[m*30:m*30+30,:,:],axis =0) 
        anomsal_30pt0_flux_stor[yr,:,m] = np.nansum(hind_month, axis = 1)
            
pickle.dump(anomsal_30pt0_flux_stor, open("anomsal_30pt0_flux_stor.pkl", 'wb'))

sal_flux_stor = np.zeros([7,40,12])

hinds = [trans_sal_2013_kg_per_gridcell,trans_sal_2014_kg_per_gridcell,trans_sal_2015_kg_per_gridcell,
         trans_sal_2016_kg_per_gridcell,trans_sal_2017_kg_per_gridcell,
        trans_sal_2018_kg_per_gridcell,trans_sal_2019_kg_per_gridcell]

for yr in range(0,7):
    hind = hinds[yr]; hind[hind == 0] = np.nan
    for m in range(0,12):
        hind_month = np.nanmean(hind[m*30:m*30+30,:,:],axis =0) 
        sal_flux_stor[yr,:,m] = np.nansum(hind_month, axis = 1)
            
pickle.dump(sal_flux_stor, open("sal_flux_stor.pkl", 'wb'))
```

```python
bath = '/data/tjarniko/MEOPAR/grid/mesh_mask201702.nc'
grid = nc.Dataset(bath)

ulength = (grid['e2u'][0,361:(361+75),23])

udep = grid['e3u_0'][0,:,361:(361+75),23]

slice_ar = np.zeros([40,75])

for i in range(0,40):
    
    slice_ar[i,:] = ulength * udep[i,:]

size_ar_masked = slice_ar*grid['umask'][0,:,361:361+75,23]
size_ar_perrow = np.nansum(size_ar_masked,axis=1)
size_ar_pr_bc12 = np.zeros([40,12])

for i in range(0,12):
    size_ar_pr_bc12[:,i] = size_ar_perrow
```
