```python

fact = 0.5
fig, axs = plt.subplots(1,1, figsize=(13*fact, 15*fact), \
                        facecolor='w', edgecolor='k')

tcm = cm.cm.matter
ind = 0
for q in range(0,550,50):
    pDIC = 1800+q
    DIC = pDIC * 1e-3
    ttemp = 7
    tsal = 32

    ####
    TA_DIC = np.arange(-100,600,10) * 1e-3
    DICr = np.zeros_like(TA_DIC);
    DICr[:] = DIC
    TAr = DIC+TA_DIC 

    Tr = np.zeros_like(DICr); Tr[:] = ttemp
    Sr = np.zeros_like(DICr); Sr[:] = tsal
    zero = np.zeros_like(DICr)
    pco2r = np.zeros_like(DICr)


    ttera_is = ttemp; tsra_psu = tsal
    tzero = 0
    tpressure = 1
    tdepths = 0

    for i in range(0,len(DICr)):
        ttara = TAr[i]
        tdra = DICr[i]


        response_tup = mocsy.mvars(temp=ttera_is, sal=tsra_psu, alk=ttara, dic=tdra, 
                           sil=tzero, phos=tzero, patm=tpressure, depth=tdepths, lat=tzero, 
                            optcon='mol/m3', optt='Tinsitu', optp='m',
                            optb = 'l10', optk1k2='m10', optkf = 'dg', optgas = 'Pinsitu')
        pH,pco2,fco2,co2,hco3,co3,OmegaA,OmegaC,BetaD,DENis,p,Tis = response_tup

        pco2r[i] = pco2[0]

    #plt.plot(TA_DIC*1e3,pco2r)


    axs.plot(TA_DIC*1e3, pco2r, color = tcm(1.*(ind)/10), label = f'DIC = {pDIC} umol/L')
    axs.legend()
    ind = ind+1
```
