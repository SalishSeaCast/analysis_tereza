```python
#(⇥max=26 !C, ⇥range=14 !C),
 
def f_t_diat(t):
    t_r = 14
    t_m = 26
    fact = 0
    if (t < (t_m- t_r)):
        fact = 1
    if (t >= (t_m- t_r)) & (t <= t_m):
        fact = (t_m-t)/t_r
    if (t > (t_m)):
        fact = 0
    
    #print(fact)
    f_t = np.exp(.07*(t -20)) *fact
    
    return f_t
```

```python
ts = np.arange(8,20,0.25)
fts = np.zeros_like(ts)
for i in range(0,len(ts)):
    tt = ts[i]
    tf_t = f_t_diat(tt)
    fts[i] = tf_t
```

```python
fig = plt.figure(figsize=(10,12))

nop = 4
for i in range(1,nop+1):
    if i == 1:
        ax = fig.add_subplot(nop,1,i)
        ax.plot(ts,fts)
        plt.title('diatom growth rate temp dependence')
    if i == 2:
        ax = fig.add_subplot(nop,1,i)
        gr = fts * 5.5e-5
        ax.plot(ts,gr)
        plt.title('diatom growth rate, no limitation by anything (L==1, upper bound)')
    if i == 3:
        ax = fig.add_subplot(nop,1,i)
        gr = fts * 5.5e-5 * 0.0106
        ax.plot(ts,gr)
        plt.title('uptake of NH4+ by diat, diat. conc. 0.0106 uM N, uptake of N03 == 0')
    if i == 4:
        ax = fig.add_subplot(nop,1,i)
        gr = fts * 5.5e-5 * 0.0106 * (106/16)
        ax.plot(ts,gr)
        plt.title('uptake of DIC by diat, diat. conc. 0.0106 uM N, uptake of N03 == 0')
        ax.set_xlabel('temp C')
```
