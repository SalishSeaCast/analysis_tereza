#### CALCULATE DEPTH-INTEGRATED BIOLOGY

```python
flag ='../pkls/BR3_flagellates_means_inletsIN.pkl'
diat ='../pkls/BR3_diatoms_means_inletsIN.pkl'
cili ='../pkls/BR3_ciliates_means_inletsIN.pkl'
csd = '../pkls/cellsize_alg2.pkl'

flag_mas = pickle.load(open(flag, 'rb'))  
diat_mas = pickle.load(open(diat, 'rb'))
cili_mas = pickle.load(open(cili, 'rb'))

mmask = nc.Dataset('/data/tjarniko/MEOPAR/grid/mesh_mask201702.nc')
depths = (mmask['e3t_1d'][:])

mask_broad = np.zeros([40,365])

for i in range(0,365):
    mask_broad[:,i] = depths


flag_mas_mm_m2 = np.nansum((mask_broad*flag_mas),axis = 0)
diat_mas_mm_m2 = np.nansum((mask_broad*diat_mas),axis = 0)
cili_mas_mm_m2 = np.nansum((mask_broad*cili_mas),axis = 0)
```

```python

def make_diffplots(y,x):
    depths = mesh['gdept_1d'][0,:]
    BR_grid_0701 = nc.Dataset('/data/tjarniko/results/BASERUN_EXP/MAIN/BR_3rd_2015/ncs/SKOG_1d_20150630_20150729_grid_T_20150701-20150701.nc')
    PI_grid_0701 = nc.Dataset('/data/tjarniko/results/BASERUN_EXP/MAIN/PI_ACBC_2015_3/ncs/SKOG_1d_20150630_20150729_grid_T_20150701-20150701.nc')

    BR_grid_0701_sal = BR_grid_0701['vosaline'][0,:,y,x]
    BR_grid_0701_sal[BR_grid_0701_sal==0] = np.nan
    BR_grid_0701_temp = BR_grid_0701['votemper'][0,:,y,x]
    BR_grid_0701_temp[BR_grid_0701_temp==0] = np.nan

    PI_grid_0701_sal = PI_grid_0701['vosaline'][0,:,y,x]
    PI_grid_0701_sal[PI_grid_0701_sal==0] = np.nan
    PI_grid_0701_temp = PI_grid_0701['votemper'][0,:,y,x]
    PI_grid_0701_temp[PI_grid_0701_temp==0] = np.nan

    PI_grid_0701_dens = gsw.rho(PI_grid_0701_sal,PI_grid_0701_temp,depths)
    BR_grid_0701_dens = gsw.rho(BR_grid_0701_sal,BR_grid_0701_temp,depths)



    BR_umolkg_0701 = pickle.load(open('../KEY_PAPERFIGURES/pkls/umolkg/BR_DIC_umolkg_20150701.pkl', 'rb'))
    BR_umolL_0701 = nc.Dataset('/data/tjarniko/results/BASERUN_EXP/MAIN/BR_3rd_2015/ncs/SKOG_1d_20150630_20150729_carp_T_20150701-20150701.nc')
    PI_umolkg_0701 = pickle.load(open('../KEY_PAPERFIGURES/pkls/umolkg/PI_DIC_umolkg_20150701.pkl', 'rb'))
    PI_umolL_0701 = nc.Dataset('/data/tjarniko/results/BASERUN_EXP/MAIN/PI_ACBC_2015_3/ncs/SKOG_1d_20150630_20150729_carp_T_20150701-20150701.nc')


    BR_umolkg_0701_DIC = BR_umolkg_0701[:,y,x]
    BR_umolkg_0701_DIC[BR_umolkg_0701_DIC==0] = np.nan
    BR_umolL_0701_DIC = BR_umolL_0701['dissolved_inorganic_carbon'][0,:,y,x]
    BR_umolL_0701_DIC[BR_umolL_0701_DIC==0] = np.nan

    PI_umolkg_0701_DIC = PI_umolkg_0701[:,y,x]
    PI_umolkg_0701_DIC[PI_umolkg_0701_DIC==0] = np.nan
    PI_umolL_0701_DIC = PI_umolL_0701['dissolved_inorganic_carbon'][0,:,y,x]
    PI_umolL_0701_DIC[PI_umolL_0701_DIC==0] = np.nan


    fact = 0.8
    fig = plt.figure(figsize=(16.5*fact, 8.5*fact))
    depth = mesh['e3t_0'][0,:,y,x]
    ax1 = plt.subplot2grid((1,3), (0,0), colspan=1, rowspan=1)
    ax1.plot(BR_umolL_0701_DIC,-depths, linestyle = '-', color = 'red', label = 'BR umol/L')
    ax1.plot(BR_umolkg_0701_DIC,-depths,linestyle = ':', color = 'red', label = 'BR umol/kg')
    ax1.plot(PI_umolL_0701_DIC,-depths, linestyle = '-', color = 'cornflowerblue', label = 'PI umol/L')
    ax1.plot(PI_umolkg_0701_DIC,-depths,linestyle = ':', color = 'cornflowerblue', label = 'PI umol/kg')
    plt.legend(loc = 'best', fontsize = 10)
    plt.xticks(rotation = 65)

    ax2 = plt.subplot2grid((1,3), (0,1), colspan=1, rowspan=1)
    ax2.plot(BR_umolL_0701_DIC-PI_umolL_0701_DIC,-depths, linestyle = '-', \
             color = 'black', label = 'BR-PI umol/L')
    ax2.plot(BR_umolkg_0701_DIC-PI_umolkg_0701_DIC,-depths, linestyle = ':', \
             color = 'black', label = 'BR-PI umol/kg')
    plt.legend(loc = 'best', fontsize = 10)
    plt.xticks(rotation = 65)

    ax3 = plt.subplot2grid((1,3), (0,2), colspan=1, rowspan=1)
    ax3.plot(PI_grid_0701_dens,-depths, linestyle = '-', \
             color = 'cornflowerblue', label = 'PI dens in-situ kg/m3')
    ax3.plot(BR_grid_0701_dens,-depths, linestyle = '-', \
             color = 'red', label = 'BR dens in-situ kg/m3')
    plt.xticks(rotation = 65)
    plt.legend(loc = 'best', fontsize = 10)

    plt.suptitle('day = 0701, x, y are:' + str(x) + ', '+str(y))
```
