```python
hind_09 = \
pickle.load(open('./hindcast_201905_timeseries_domain_undersat/hind_undersat_2009.pkl', 'rb'))   
hind_10 = \
pickle.load(open('./hindcast_201905_timeseries_domain_undersat/hind_undersat_2010.pkl', 'rb'))
hind_11 = \
pickle.load(open('./hindcast_201905_timeseries_domain_undersat/hind_undersat_2011.pkl', 'rb'))   
hind_13 = \
pickle.load(open('./hindcast_201905_timeseries_domain_undersat/hind_undersat_2013.pkl', 'rb'))   
hind_14 = \
pickle.load(open('./hindcast_201905_timeseries_domain_undersat/hind_undersat_2014.pkl', 'rb'))   
hind_15 = \
pickle.load(open('./hindcast_201905_timeseries_domain_undersat/hind_undersat_2015.pkl', 'rb'))   

```

```python
under_threshold = [0.5,0.6,0.7,0.8,0.9,1.0,1.1,1.2,1.3,1.4,1.5]

hinds = [hind_09,hind_10,hind_11,hind_13,hind_14,hind_15]
years = ['2009','2010','2011','2013','2014','2015']
fs = 18
fact = 0.7
lw = 2.5
days = np.arange(1,366,1)
tcm = cm.cm.phase
```

```python
thres_ind = 0
thres = under_threshold[thres_ind]

fig = plt.figure(figsize=(14*fact,6*fact))

ax1 = fig.add_subplot(1,1,1)

for i in range(0,len(hinds)):
    t_hind = hinds[i]
    yr = years[i]
    
    ax1.plot(days,t_hind[thres_ind,:]*100,color = tcm(1.*i/6),\
             linewidth = lw, linestyle = '-', label = yr)

ax1.tick_params(axis='both', which='major', labelsize=15)
ax1.legend(bbox_to_anchor=[1, 1], fontsize = fs-4.5)

w = ax1.set_xticks(np.arange(0, 365, step=30))
ax1.set_xlim(0,365)
w = ax1.set_yticks(np.arange(0, 1, step=.1))
ax1.set_ylabel('% domain volume below threshold', fontsize = 18)
ax1.set_title('% domain volume below $\Omega_A$ thres '+str(thres), fontsize = 18)
ax1.set_xlabel('days', fontsize = 18)

```
