#### Figure 2 - carbon balance comparison - 3 figures

```python
fx = 13
fy = 8
t = np.arange(0,365)
t_sig = np.sin(t)

fig = plt.figure(figsize=(fx,fy))

ax = fig.add_subplot(3,1,1)
ax.plot(t, t_sig, color = 'red',linewidth = 2 , label = 'BR')
ax.plot(t, t_sig-0.5, color = 'blue',linewidth = 2 , label = 'PI_AC')
ax.plot(t, t_sig-0.75, color = 'blue',linewidth = 2 , linestyle =':', label = 'PI_ACBC')
ax.plot(t, t_sig+0.5, color = 'green',linewidth = 2 , label = 'CP_AC')
ax.plot(t, t_sig+0.75, color = 'green',linewidth = 2 ,  linestyle =':', label = 'PI_ACBC')
ax.set_title('Cumulative domain carbon')
ax.legend()
ax.set_xlim([1,365])

ax = fig.add_subplot(3,1,2)
ax.plot(t, t_sig, color = 'red',linewidth = 2 , label = 'BR')
ax.plot(t, t_sig-0.5, color = 'blue',linewidth = 2 , label = 'PI_AC')
ax.plot(t, t_sig-0.75, color = 'blue',linewidth = 2 , linestyle =':', label = 'PI_ACBC')
ax.plot(t, t_sig+0.5, color = 'green',linewidth = 2 , label = 'CP_AC')
ax.plot(t, t_sig+0.75, color = 'green',linewidth = 2 ,  linestyle =':', label = 'PI_ACBC')
ax.set_title('Cumulative surface-layer [20m] carbon')
ax.legend()
ax.set_xlim([1,365])

ax = fig.add_subplot(3,1,3)
ax.plot(t, t_sig, color = 'red',linewidth = 2 , label = 'BR')
ax.plot(t, t_sig-0.5, color = 'blue',linewidth = 2 , label = 'PI_AC')
ax.plot(t, t_sig-0.75, color = 'blue',linewidth = 2 , linestyle =':', label = 'PI_ACBC')
ax.plot(t, t_sig+0.5, color = 'green',linewidth = 2 , label = 'CP_AC')
ax.plot(t, t_sig+0.75, color = 'green',linewidth = 2 ,  linestyle =':', label = 'PI_ACBC')
ax.set_title('Cumulative deep [20m] carbon')
ax.legend()
ax.set_xlim([1,365])

ax.set_xlim([1,365])
```
